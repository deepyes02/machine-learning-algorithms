### Linear Regression Theory ###
* Here, we have a dataset that explains how many miles per gallon a car can go and the weight of the car. Based on the dataset, we will establish the relationship between car's weight and miles per gallon it goes, other things remaining the same.
* Based on this dataset, we will predict miles per gallon for other cars based on their weight.
* Based on these excercises, we will create a simple model and understand the terms of linear regression algorithm in Machine learning.


**Linear Regression Equation**  

y = mx + b  
$y^l = w_1x_1$ + b  
where, 
* b = bias of model / y-intercept
* $w_1$ = weight of the feature / slope (m)
* $x_1$ = the input

**NOTE**  
bias $b$ and weight $w$ are calculated during training  

**If a model has multiple features**  
Then the linear regression equation can be written as:  

$y^l = b + w_1x_1 + w_2x_2 + w_3x3 .... + w_nx_n$  
where,  
* x1,x2,x3,etc are inputs like pounds, displacement, acceleration, engine-size and so on.  

```python
import pandas as pd
import seaborn as sns
data = {
    'Pounds in 1000s': [3.5, 3.69, 3.44, 3.43, 4.34, 4.42, 2.37],
    'Miles per gallon': [18, 15, 18, 16, 15, 14, 24]
}
df = pd.DataFrame(data)

g = sns.scatterplot(x='Pounds in 1000s', y='Miles per gallon', data=df)
g= sns.regplot(x='Pounds in 1000s', y='Miles per gallon', data=df, ci=None, label="Some data", line_kws={'linewidth':1}, scatter_kws={'s': 2})
g.set(ylim=(12, 25))
g.set(xlim=(2, 5))
```

## LOSS ##  
Loss is a numerical metric that calculates how wrong a model's predictions are.  

There are four types of loss  

| Loss type | Definition | Equation | 
|----|-----|---|
| $L_1$ loss | Sum of absolute difference between predicted value and actual value | $\sum{\|\ actual value - predicted value\ \|}$ |
| Mean Absolute Error (MAE) | Average of $L_1$ losses across the dataset | $ \frac{1}{N} \sum{\|\ actual value - predicted value\| }$ |
| $L_2$ loss | Sum of squared differences between predicted value and actual value | $ \sum{(\ actual value - predicted value)^2 }$ |
| Mean Squared Error (MSE) | Average of $L_1$ losses across the dataset | $ \frac{1}{N} \sum{(\ actual value - predicted value)^2 }$ |
