### Linear Regression Theory ###
* Here, we have a dataset that explains how many miles per gallon a car can go and the weight of the car. Based on the dataset, we will establish the relationship between car's weight and miles per gallon it goes, other things remaining the same.
* Based on this dataset, we will predict miles per gallon for other cars based on their weight.
* Based on these excercises, we will create a simple model and understand the terms of linear regression algorithm in Machine learning.

```python
import pandas as pd
import seaborn as sns
data = {
    'Pounds in 1000s': [3.5, 3.69, 3.44, 3.43, 4.34, 4.42, 2.37],
    'Miles per gallon': [18, 15, 18, 16, 15, 14, 24]
}
df = pd.DataFrame(data)

sns.scatterplot(x='Pounds in 1000s', y='Miles per gallon', data=df)

```
